# criação de agrupamentos Português 7 REG1



```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r'arquivos/avaliacao-port.csv', encoding= 'latin-1', sep= ';')
df = df.dropna()
display(df)
display(df.describe())
```

```python
# === Matriz dos atributos ===

X = df.drop('Student Name', axis = 1)


# == rodar o KMeans (padronizado)

scaler = StandardScaler()
X_padronizado = scaler.fit_transform(X)

kmeans = KMeans(n_clusters= 8 , random_state= 0 , n_init= 20)

df['agrupamento'] = kmeans.fit_predict(X_padronizado)



```

```python
agrupamentos_aluno_df = df[['Student Name', 'agrupamento']].groupby('agrupamento').count()
agrupamentos_aluno_df
```

```python
agrupamentos_hab_df = df.drop('Student Name', axis = 1).groupby('agrupamento').mean()
agrupamentos_hab_df
```
