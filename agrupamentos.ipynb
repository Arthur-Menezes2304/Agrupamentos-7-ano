# Aqui, irei fazer um tratamento de colunas para o exame de português.

### Existem várias colunas repetidas, então eu preciso juntá-las somando os seus valores primeiro antes de aplicar o algoritmo.

```python
# lendo os arquivos
df = pd.read_csv(r'arquivos/avaliacao-port-reg1.csv', encoding= 'latin-1', sep= ';')
df = df.dropna()

# Passo 1: Criar um dicionário agrupando colunas pelo nome base
from collections import defaultdict
# remove o sufixo .1, .2 etc
col_groups = defaultdict(list)
for col in df.columns:
    base = col.split('.')[0]
    col_groups[base].append(col)

# Passo 2: Somar as colunas repetidas
df_combined = pd.DataFrame()

for base, cols in col_groups.items():
    if len(cols) == 1:
        df_combined[base] = df[cols[0]]
    else:
        df_combined[base] = df[cols].sum(axis=1, skipna=True)



colunas = df_combined.columns
# alterando os valores para inteiro
for i in range(1, len(colunas) -1):
    df_combined[colunas[i]] = df_combined[colunas[i]].astype('int64')


# df_combined agora tem colunas únicas e somadas
df = df_combined
df
```

```python
# === Matriz dos atributos ===

X = df.drop(['Student Name', 'Score (#)','TURMA'], axis = 1)


# == rodar o KMeans (padronizado)

scaler = StandardScaler()
X_padronizado = scaler.fit_transform(X)

kmeans = KMeans(n_clusters= 8, random_state= 0 , n_init= 50)

df['agrupamento'] = kmeans.fit_predict(X_padronizado)



```

```python
df.info()
```

```python
agrupamentos_aluno_df = df[['Student Name', 'agrupamento']].groupby('agrupamento').count()
agrupamentos_aluno_df
```

### Renumerar os agrupamentos 

```python
df['agrupamento'] = df['agrupamento'] + 1
```

### para não ficar com grupos muito grandes, vamos distribuir estudantes do grupo que possui mais para os grupos que possuem menos. Assim, deixamos um pouco mais equilibrado

```python
agrupamentos_hab_df = df.drop(['Student Name', 'Score (#)','TURMA'], axis = 1).groupby('agrupamento').mean()
agrupamentos_hab_df
```
